```python
import os
import re
from typing import List
from pathlib import Path
from langchain.docstore.document import Document
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import MarkdownHeaderTextSplitter
```

## 1. ë§ˆí¬ë‹¤ìš´ ë¬¸ì„œ ì²˜ë¦¬

```python
def hierarchical_markdown_split(md_text: str, path_prefix: str = "") -> list[Document]:
    """ë§ˆí¬ë‹¤ìš´ ë¬¸ì„œë¥¼ ê³„ì¸µì ìœ¼ë¡œ ë¶„í• í•©ë‹ˆë‹¤."""
    splitter = MarkdownHeaderTextSplitter(headers_to_split_on=[
        ("#", "title"),
        ("##", "section"),
        ("###", "subsection"),
        ("####", "subsubsection")
    ])
    docs = splitter.split_text(md_text)

    result_docs = []
    current_title = None
    for doc in docs:
        metadata = doc.metadata
        if "title" in metadata:
            current_title = metadata["title"]

        if current_title:
            full_title = ""
            if "subsection" in metadata:
                full_title += f" > {metadata['subsection']}"
            if "subsubsection" in metadata:
                full_title += f" > {metadata['subsubsection']}"

            content = f"[{full_title}]\n\n{doc.page_content}"
            doc = Document(page_content=content, metadata={
                **doc.metadata,
                "type": "documentation",
                "source": "dev_center_guide_touched.md"
            })

        result_docs.append(doc)

    return result_docs

def load_markdown_file(file_path: str) -> str:
    """ë§ˆí¬ë‹¤ìš´ íŒŒì¼ì„ ë¡œë“œí•©ë‹ˆë‹¤."""
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# ë§ˆí¬ë‹¤ìš´ íŒŒì¼ ë¡œë“œ ë° ë¶„í• 
str_md_file = load_markdown_file("data/dev_center_guide_touched.md")
docs_markdown = hierarchical_markdown_split(str_md_file)

print(f"ë§ˆí¬ë‹¤ìš´ ë¬¸ì„œ ë¶„í•  ì™„ë£Œ: {len(docs_markdown)}ê°œ ì²­í¬")
```

## 2. Kotlin ì½”ë“œ ì²˜ë¦¬

```python
def extract_package(content: str) -> str:
    """Kotlin íŒŒì¼ì—ì„œ íŒ¨í‚¤ì§€ ì„ ì–¸ì„ ì¶”ì¶œí•©ë‹ˆë‹¤."""
    package_match = re.search(r'package\s+([\w.]+)', content)
    return package_match.group(1) if package_match else ""

def extract_class_blocks(content: str) -> List[tuple]:
    """Kotlin íŒŒì¼ì—ì„œ í´ë˜ìŠ¤ ë¸”ë¡ì„ ì¶”ì¶œí•©ë‹ˆë‹¤."""
    blocks = []
    
    # í´ë˜ìŠ¤, ê°ì²´, ì¸í„°í˜ì´ìŠ¤, í•¨ìˆ˜ ë“± ì¶”ì¶œ
    patterns = [
        (r'(class\s+\w+[\s\S]*?})', 'class'),
        (r'(object\s+\w+[\s\S]*?})', 'object'),
        (r'(interface\s+\w+[\s\S]*?})', 'interface'),
        (r'(fun\s+\w+[\s\S]*?})', 'function'),
        (r'(companion\s+object[\s\S]*?})', 'companion_object')
    ]
    
    for pattern, symbol_type in patterns:
        matches = re.finditer(pattern, content, re.MULTILINE)
        for match in matches:
            block_text = match.group(1)
            # í´ë˜ìŠ¤/í•¨ìˆ˜ ì´ë¦„ ì¶”ì¶œ
            name_match = re.search(r'(class|object|interface|fun|companion\s+object)\s+(\w+)', block_text)
            symbol_name = name_match.group(2) if name_match else f"{symbol_type}_{len(blocks)}"
            
            blocks.append((f"{symbol_type}:{symbol_name}", block_text))
    
    return blocks

def load_kotlin_documents(project_root: str) -> list[Document]:
    """Kotlin í”„ë¡œì íŠ¸ì—ì„œ ë¬¸ì„œë¥¼ ë¡œë“œí•©ë‹ˆë‹¤."""
    project_root_path = Path(project_root)
    documents = []

    for file_path in project_root_path.rglob("*.kt"):
        rel_path = file_path.relative_to(project_root_path)
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()

        package = extract_package(content)
        blocks = extract_class_blocks(content)
        
        for symbol, block in blocks:
            doc = Document(
                page_content=block,
                metadata={
                    "source": str(rel_path),
                    "symbol": symbol,
                    "package": package,
                    "type": "code",
                    "language": "kotlin"
                }
            )
            documents.append(doc)

    return documents

# Kotlin ì½”ë“œ ë¡œë“œ
docs_kotlin = load_kotlin_documents("code_sample/onestore_iap_release/")
print(f"Kotlin ì½”ë“œ ë¬¸ì„œ ë¡œë“œ ì™„ë£Œ: {len(docs_kotlin)}ê°œ ì²­í¬")

# ìƒ˜í”Œ ì¶œë ¥
if docs_kotlin:
    print("\nì²« ë²ˆì§¸ ì½”ë“œ ë¬¸ì„œ ìƒ˜í”Œ:")
    print(f"ë©”íƒ€ë°ì´í„°: {docs_kotlin[0].metadata}")
    print(f"ë‚´ìš© (ì²˜ìŒ 200ì): {docs_kotlin[0].page_content[:200]}...")
```

## 3. ë¬¸ì„œ í†µí•© ë° ì„ë² ë”© ìƒì„±

```python
def embed_and_save(docs: List[Document], output_path: str):
    """ë¬¸ì„œë¥¼ ì„ë² ë”©í•˜ê³  FAISS ë°ì´í„°ë² ì´ìŠ¤ë¡œ ì €ì¥í•©ë‹ˆë‹¤."""
    # ì„ë² ë”© ëª¨ë¸ ì´ˆê¸°í™”
    embedding_model = OllamaEmbeddings(model="exaone3.5:latest")
    
    # FAISS ë°ì´í„°ë² ì´ìŠ¤ ìƒì„± ë° ì €ì¥
    db = FAISS.from_documents(docs, embedding_model)
    db.save_local(output_path)
    print(f"âœ… ì„ë² ë”© ì €ì¥ ì™„ë£Œ: {output_path}")

# ëª¨ë“  ë¬¸ì„œ í†µí•©
total_docs = docs_markdown + docs_kotlin
print(f"ì´ {len(total_docs)}ê°œì˜ ë¬¸ì„œë¥¼ ìƒì„±í•˜ì˜€ìŠµë‹ˆë‹¤.")
print(f"- ë§ˆí¬ë‹¤ìš´ ë¬¸ì„œ: {len(docs_markdown)}ê°œ")
print(f"- Kotlin ì½”ë“œ: {len(docs_kotlin)}ê°œ")

# ì„ë² ë”© ìƒì„± ë° ì €ì¥
output_dir = "models/faiss_vs_rag_iap_v7x"
os.makedirs(output_dir, exist_ok=True)
embed_and_save(total_docs, output_dir)
```

## 4. ì„ë² ë”© ê²€ì¦

```python
# ì €ì¥ëœ ì„ë² ë”© ë¡œë“œ ë° í…ŒìŠ¤íŠ¸
embedding_model = OllamaEmbeddings(model="exaone3.5:latest")
loaded_db = FAISS.load_local(
    folder_path=output_dir,
    embeddings=embedding_model,
    allow_dangerous_deserialization=True,
)

# ê²€ìƒ‰ í…ŒìŠ¤íŠ¸
test_queries = [
    "ì¸ì•±ê²°ì œ ì´ˆê¸°í™” ë°©ë²•",
    "ì •ê¸°ê²°ì œ êµ¬í˜„ ì½”ë“œ",
    "PNS ì„¤ì • ë°©ë²•",
    "ê²°ì œ ê²€ì¦ API"
]

for query in test_queries:
    print(f"\nğŸ” ê²€ìƒ‰ ì¿¼ë¦¬: {query}")
    results = loaded_db.similarity_search(query, k=3)
    
    for i, doc in enumerate(results, 1):
        print(f"\nê²°ê³¼ {i}:")
        print(f"íƒ€ì…: {doc.metadata.get('type', 'unknown')}")
        print(f"ì†ŒìŠ¤: {doc.metadata.get('source', 'unknown')}")
        print(f"ë‚´ìš©: {doc.page_content[:200]}...")
        print("-" * 50)
```

## 6. RAG ì²´ì¸ ìƒì„± ë° ì§ˆë¬¸ ì‘ë‹µ

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

# RAG ì²´ì¸ êµ¬ì„± ìš”ì†Œ
embedding_model = OllamaEmbeddings(model="exaone3.5:latest")
llm = ChatOllama(model="exaone3.5:latest", temperature=0.3)

# ê²€ìƒ‰ê¸° ì„¤ì •
retriever = loaded_db.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 8, "fetch_k": 20, "lambda_mult": 0.6}
)

# í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿
prompt_template = """
ë‹¹ì‹ ì€ ì›ìŠ¤í† ì–´ ì¸ì•±ê²°ì œ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ì£¼ì–´ì§„ ì»¨í…ìŠ¤íŠ¸ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì§ˆë¬¸ì— ë‹µë³€í•´ì£¼ì„¸ìš”.

ë‹µë³€ ì‹œ ë‹¤ìŒ ì‚¬í•­ì„ ê³ ë ¤í•´ì£¼ì„¸ìš”:
1. í•œêµ­ì–´ë¡œ ëª…í™•í•˜ê³  ì´í•´í•˜ê¸° ì‰½ê²Œ ë‹µë³€í•˜ì„¸ìš”
2. ì½”ë“œ ì˜ˆì‹œê°€ ìˆë‹¤ë©´ í¬í•¨í•´ì£¼ì„¸ìš”
3. ë‹¨ê³„ë³„ë¡œ ì„¤ëª…í•´ì£¼ì„¸ìš”
4. ê°œë°œì ê´€ì ì—ì„œ ì‹¤ìš©ì ì¸ ì •ë³´ë¥¼ ì œê³µí•´ì£¼ì„¸ìš”
5. ì»¨í…ìŠ¤íŠ¸ì— ì—†ëŠ” ë‚´ìš©ì€ "í•´ë‹¹ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤"ë¼ê³  ë‹µë³€í•˜ì„¸ìš”

ì»¨í…ìŠ¤íŠ¸:
{context}

ì§ˆë¬¸: {question}

ë‹µë³€:"""

prompt = PromptTemplate.from_template(prompt_template)

# RAG ì²´ì¸ êµ¬ì„±
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

print("âœ… RAG ì²´ì¸ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
```
